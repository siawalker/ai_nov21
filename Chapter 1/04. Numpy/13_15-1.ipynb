```python
#create 10x10 array with random values, find minimum and maximum
import numpy as np
a = np.random.random((10,10))
print(a)
amin, amax = a.min(), a.max()
print("Minimum and Maximum values:")
print(amin, amax)
```

```python
# create a 2nd array with 1 on borfder and zero inside
import numpy as np
a = np.ones((10,10))
print(a)
print("1 in the border and 0 in the array")
a[1:-1,1:-1]= 0
print(a)
```
